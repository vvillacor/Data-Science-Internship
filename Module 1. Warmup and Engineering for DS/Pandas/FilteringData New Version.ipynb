<font size ="10">Filtering and Sorting Data (corrections) <size>

```python
import pandas as pd
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipo = pd.read_csv(url, sep = '\t')
```

```python
#chipo.head(15)
chipo.item_name.unique()
```

<font size ="5">Step 4. How many products cost more than $10.00? <size>
    

```python
chipo.item_price = chipo.item_price.apply(lambda x: float(x[1:-1]))
```

```python
# products offered by Chipo, not number of items sold that cost over $10
chipo_filtered = chipo.drop_duplicates(['item_name', 'choice_description', 'item_price'])
products_over_10 = chipo_filtered[chipo_filtered['item_price'] > 10.00]
print("There are " + str(products_over_10.shape[0]) + " products which price is over 10.00.")
```

<font size ="5">Step 7. What was the quantity of the most expensive item ordered?<size>

```python
chipo.loc[chipo["item_price"].idxmax()]["quantity"]
```

```python
#This is returning the same answer from the old notebook, might ask chatgpt what is going on...  RESPONSE:
# Your logic is impeccable for finding the index of the maximum value and then querying the quantity in that same position.
# However, there's a technical "trap" in this specific dataset that could give you an incorrect answer. 
#To answer "What was the quantity ordered of the most expensive item?", we first need to identify which item has the highest unit
#price and then see how much of it was ordered.
chipo['unit_price'] = chipo['item_price'] / chipo['quantity']
max_unit_price = chipo['unit_price'].max()
most_expensive_items = chipo[chipo['unit_price'] == max_unit_price]
total_quantity = most_expensive_items.quantity.sum()
print(total_quantity)
```
