Step 5. Discover the Male ratio per occupation and sort it from the most to the least</font>

```python
users['gender_n'] = users['gender'].apply(lambda x: 1 if x == 'M' else 0)

# 2. Agrupamos por ocupaci√≥n, calculamos la media y multiplicamos por 100
# 3. Ordenamos de mayor a menor (ascending=False)
male_ratio = (users.groupby('occupation').gender_n.mean() * 100).sort_values(ascending=False)

print(male_ratio)
```

<font size ="5">Step 8. For each occupation present the percentage of women and men</font>

```python
users['is_male'] = users['gender'].apply(lambda x: 1 if x == 'M' else 0) 
male_ratio = users.groupby('occupation').is_male.mean() * 100
female_ratio = 100 - male_ratio
gender_perc = pd.DataFrame({
    'Women %': female_ratio,
    'Men %': male_ratio
})
```
