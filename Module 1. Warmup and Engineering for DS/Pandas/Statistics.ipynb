<font size="14">US - Baby Names </font>

```python
import pandas as pd
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv"
baby_names = pd.read_csv(url)
```

Step 4. See the first 10 entries

```python
baby_names.head(10)
```

<font size="5">Step 5. Delete the column 'Unnamed: 0' and 'Id'</font>

```python
baby_names=baby_names.drop(columns=["Unnamed: 0","Id"])
```

<font size="5">Step 6. What year has the highest number of baby names in the dataset?
</font>

```python
names_per_year = baby_names.groupby('Year')['Count'].sum()
best_year = names_per_year.idxmax()
print(best_year)
```

<font size= "5">Step 7. Is there more male or female names in the dataset? <font>

```python
print(baby_names['Gender'].value_counts())
```

<font size="5">Step 11. How many different names have the least occurrences?
 </font>

```python
least_oc = baby_names.Name.value_counts().tail(3685) 
print(least_oc)
```

```python
names_total = baby_names.groupby('Name')['Count'].sum()
min_value = names_total.min()
names_with_min_occ = len(names_total[names_total == min_value])
print(f"Hay {names_with_min_occ} nombres diferentes con solo {min_value} ocurrencias.")
```

<font size="5">Step 12. What is the median name occurrence?
 </font>

```python
median_occ = names_total.median()
print(median_occ)
```

<font size="5">Step 13. What is the standard deviation of names?<font>

```python
std_occ = names_total.std()
print(std_occ)
```
