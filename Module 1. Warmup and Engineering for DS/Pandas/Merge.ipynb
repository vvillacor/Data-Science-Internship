<font size="14">MPG Cars </font>

```python
import pandas as pd
import numpy as np
```

```python
url1 = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars1.csv"
url2 =  "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars2.csv"
cars1 = pd.read_csv(url1)
cars2 = pd.read_csv(url2)
```

<font size="5">Step 4. Oops, it seems our first dataset has some unnamed blank columns, fix cars1 </font>

```python
#cars1 = cars1.drop(columns=['Unnamed: 9', 'Unnamed: 10',
 #'Unnamed: 11', 'Unnamed: 12', 'Unnamed: 13'])
# this is not working
cars1 = cars1.dropna(axis=1, how='all')
cars1.head()
```

<font size="5">Step 5. What is the number of observations in each dataset?</font>

```python
print(len(cars1))
print(len(cars2))
```

<font size="5">Step 6. Join cars1 and cars2 into a single DataFrame called cars </font>

```python
cars = pd.concat([cars1, cars2], ignore_index=True)
```
