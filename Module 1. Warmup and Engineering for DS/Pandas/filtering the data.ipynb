<font size ="8"> Filtering and Sorting Data</font>

```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
```

```python
chipo = pd.read_csv(url, sep='\t')
```

```python
chipo.head()
```

<font size ="5">Step 4. How many products cost more than $10.00?</font>

```python
products_over_10 = chipo.loc[chipo.item_price>=10.00]
```

```python
#  data type change is needed here 
```

```python
chipo.item_price = chipo.item_price.apply(lambda x: float(x[1:-1]))
```

```python
chipo.item_price.dtype
```

```python
products_over_10 = chipo.loc[chipo.item_price>=10.00]
```

```python
print("There are " + str(products_over_10.item_name.nunique()) + " products which price is over 10.00.")
```

<font size="5"> Step 5. What is the price of each item? </font>
<font size="2"> print a data frame with only three columns item_name choice_description and product_price </font>

```python
#prices = chipo.loc[:,["item_name","choice_description","item_price"]]
chipo['product_price'] = chipo['item_price'] / chipo['quantity']
prices = chipo.loc[:, ["item_name", "choice_description", "product_price"]].drop_duplicates()
```

```python
print(prices.head(10))
```

<font size="5"> Step 6. Sort by the name of the item </font>

```python
prices.sort_values(by='item_name', ascending=False)
```

<font size="5">Step 7. What was the quantity of the most expensive item ordered?</font>

```python
max_price = chipo.item_price.max()
```

```python
most_exp = chipo[chipo.item_price == max_price]
```

```python
print(most_exp)
```

<font size="5">Step 8. How many times was a Veggie Salad Bowl ordered?</font>

```python
veggie_salad_ordered_n_times = len(chipo[chipo.item_name == "Veggie Salad Bowl"])
```

```python
print("The Veggie Salad Bowl was ordered " + str(veggie_salad_ordered_n_times) + " times")
```
