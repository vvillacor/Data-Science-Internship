60. How to tell if a given 2D array has null columns? (★★☆)

```python
Z = np.array([
    [1, 0, 3],
    [4, 0, 6],
    [7, 0, 9]
])
null_columns = (~Z.any(axis=0))

# To get a simple True/False if any such column exists:
has_null_column = null_columns.any()

print(f"Column-wise null check: {null_columns}")
print(f"Does it have any null columns? {has_null_column}")
```

62. Considering two arrays with shape (1,3) and (3,1), how to compute their sum using an iterator? (★★☆)

```python
uno = np.array([1,3])
dos = np.array([3,1])
it = np.nditer([uno, dos, None])
for x,y,z in it:
    z[...] = x+y
result = it.operands[2]    
print(result)
```

70. Consider the vector [1, 2, 3, 4, 5], how to build a new vector with 3 consecutive zeros interleaved between each value? (★★★)

```python
def interleave_zeros(vector, n_zeros):
    res = np.zeros(len(vector) + (len(vector)-1) * n_zeros)
    res[::n_zeros+1] = vector
    return res
vector = np.array([1, 2, 3, 4, 5])
print(interleave_zeros(vector,3))
    
```

78. Consider 2 sets of points P0,P1 describing lines (2d) and a point p, how to compute distance from p to each line i (P0[i],P1[i])? (★★★)

```python
P0 = np.random.uniform(-10,10,(10,2))
P1 = np.random.uniform(-10,10,(10,2))
p  = np.random.uniform(-10,10,( 1,2))

def distance_faster(P0,P1,p):
    v = P1- P0 
    v[:,[0,1]] = v[:,[1,0]]
    v[:,1]*=-1 
    norm = np.linalg.norm(v,axis=1)   
    r = P0 - p
    d = np.abs(np.einsum("ij,ij->i",r,v)) / norm 

    return d

print(distance_faster(P0, P1, p))
```
